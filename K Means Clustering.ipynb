```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
cent=np.array([[1,1],[4,1]])
```

```python
kmeans=KMeans(n_clusters=2,init=cent,n_init=1)
```

```python
X=[[1,1],[2,3],[4,1],[4,3],[5,2]]
kmeans.fit(X)
```

```python
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
kmeans.score(X)
```

```python
def euc_dist(x,y):
    sumxy=0
    for i in range(len(x)):
        sumxy+=(x[i]-y[i])**2
    sumxy=sumxy**0.5
    return sumxy
my_score=0
for i in range(len(X)):
    my_score+=(euc_dist(np.array(X[i]),kmeans.cluster_centers_[kmeans.labels_[i]]))**2
print(my_score)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
X, y = make_blobs(random_state = 2)
#plt.scatter(X[:,0],X[:,1])
#plt.show()
cent=np.array([[-2,-8],[2,-5],[-3,-2]])
plt.plot(X[:, 0], X[:, 1], "co")
plt.plot(cent[0][0],cent[0][1],"co",markeredgecolor='black')
plt.plot(cent[1][0],cent[1][1],"yo",markeredgecolor='black')
plt.plot(cent[2][0],cent[2][1],"mo",markeredgecolor='black')
plt.ylabel("")
plt.xlabel("")
plt.show()
```

```python
testlabel=[]
for x in X:
    newray=[euc_dist(x,cent[0]),euc_dist(x,cent[1]),euc_dist(x,cent[2])]
    testlabel.append(newray.index(min(newray)))
testlabel=np.array(testlabel)
```

```python
fig, axes = plt.subplots(ncols=1, nrows=4, figsize=(5, 15), sharey=True, sharex=True)

plt.sca(axes[0])
plt.plot(X[:, 0][testlabel==0], X[:, 1][testlabel==0], "bs")
plt.plot(X[:, 0][testlabel==1], X[:, 1][testlabel==1], "g^")
plt.plot(X[:, 0][testlabel==2], X[:, 1][testlabel==2], "ro")
plt.plot(cent[0][0],cent[0][1],"co",markeredgecolor='black',label="old center")
plt.plot(cent[1][0],cent[1][1],"yo",markeredgecolor='black')
plt.plot(cent[2][0],cent[2][1],"mo",markeredgecolor='black')
Xnew=[np.mean(X[:,0][testlabel==0]),np.mean(X[:,1][testlabel==0])]
Ynew=[np.mean(X[:,0][testlabel==1]),np.mean(X[:,1][testlabel==1])]
Znew=[np.mean(X[:,0][testlabel==2]),np.mean(X[:,1][testlabel==2])]
plt.plot(Xnew[0],Xnew[1],"cs",markeredgecolor='black',label="new center")
plt.plot(Ynew[0],Ynew[1],"ys",markeredgecolor='black')
plt.plot(Znew[0],Znew[1],"ms",markeredgecolor='black')
plt.legend(loc="lower right")
plt.title(r"Step 0", fontsize=16)
#plt.show()

newcent=cent
for i in range(3):
    plt.sca(axes[(i+1)])
    kmeans = KMeans(n_clusters = 3,max_iter=1,init=newcent,n_init=1)
    kmeans.fit(X)
    plt.plot(X[:, 0][kmeans.labels_==0], X[:, 1][kmeans.labels_==0], "bs")
    plt.plot(X[:, 0][kmeans.labels_==1], X[:, 1][kmeans.labels_==1], "g^")
    plt.plot(X[:, 0][kmeans.labels_==2], X[:, 1][kmeans.labels_==2], "ro")
    plt.plot(kmeans.cluster_centers_[0][0],kmeans.cluster_centers_[0][1],"co",markeredgecolor='black',label="old center")
    plt.plot(kmeans.cluster_centers_[1][0],kmeans.cluster_centers_[1][1],"yo",markeredgecolor='black')
    plt.plot(kmeans.cluster_centers_[2][0],kmeans.cluster_centers_[2][1],"mo",markeredgecolor='black')
    
    Xnew=[np.mean(X[:,0][kmeans.labels_==0]),np.mean(X[:,1][kmeans.labels_==0])]
    Ynew=[np.mean(X[:,0][kmeans.labels_==1]),np.mean(X[:,1][kmeans.labels_==1])]
    Znew=[np.mean(X[:,0][kmeans.labels_==2]),np.mean(X[:,1][kmeans.labels_==2])]
    plt.plot(Xnew[0],Xnew[1],"cs",markeredgecolor='black',label="new center")
    plt.plot(Ynew[0],Ynew[1],"ys",markeredgecolor='black')
    plt.plot(Znew[0],Znew[1],"ms",markeredgecolor='black')
    plt.legend(loc="lower right")
    plt.title(r"Step ${}$".format(i+1), fontsize=16)
    #plt.show()
    
    newcent=kmeans.cluster_centers_
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
X, y = make_blobs(random_state = 2)
#plt.scatter(X[:,0],X[:,1])
#plt.show()
cent4=np.array([[-2,-8],[2,-5],[-3,-2],[0,-6]])
plt.plot(X[:, 0], X[:, 1], "co")
plt.plot(cent4[0][0],cent4[0][1],"bo",markeredgecolor='black')
plt.plot(cent4[1][0],cent4[1][1],"go",markeredgecolor='black')
plt.plot(cent4[2][0],cent4[2][1],"ro",markeredgecolor='black')
plt.plot(cent4[3][0],cent4[3][1],"wo",markeredgecolor='black')
plt.show()
```

```python
testlabel=[]
for x in X:
    newray=[euc_dist(x,cent4[0]),euc_dist(x,cent4[1]),euc_dist(x,cent4[2]),euc_dist(x,cent4[3])]
    testlabel.append(newray.index(min(newray)))
testlabel=np.array(testlabel)

things=["bs","g^","ro","kp"]
things2=["co","yo","mo","wo"]
things3=["cs","ys","ms","ws"]

Xnewray=[[],[],[],[]]
for j in range(4):
    plt.plot(X[:, 0][testlabel==j], X[:, 1][testlabel==j], things[j])
    plt.plot(cent4[j][0],cent4[j][1],things2[j],markeredgecolor='black')
    Xnewray[j]=[np.mean(X[:,0][testlabel==j]),np.mean(X[:,1][testlabel==j])]
    plt.plot(Xnewray[j][0],Xnewray[j][1],things3[j],markeredgecolor='black') 
    plt.title(r"Step 0", fontsize=16)
plt.show()

newcent=cent4
for i in range(7):
    kmeans = KMeans(n_clusters = 4,max_iter=1,init=newcent,n_init=1)
    kmeans.fit(X)
    Xnewray=[[],[],[],[]]
    for j in range(4):
        plt.plot(X[:, 0][kmeans.labels_==j], X[:, 1][kmeans.labels_==j], things[j])
        plt.plot(kmeans.cluster_centers_[j][0],kmeans.cluster_centers_[j][1],things2[j],markeredgecolor='black',label="old center")
        Xnewray[j]=[np.mean(X[:,0][kmeans.labels_==j]),np.mean(X[:,1][kmeans.labels_==j])]
        plt.plot(Xnewray[j][0],Xnewray[j][1],things3[j],markeredgecolor='black')    
    plt.title(r"Step ${}$".format(i+1), fontsize=16)
    plt.show()
    newcent=kmeans.cluster_centers_
```
